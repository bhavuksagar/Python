```python
#positional arguments-> when we have multiple parameter to pass in function, so instead of variable we can use positional argument 


def printvalue(*args):
    for i in args:
        print(i)

printvalue(1,2,3,45,6,7,8,"Bhavuk","Sagar",121,34,5,456,5,756)
```

```python
#keyword-> when we have to pass multiple key value pairs in the function **kwargs

def keyvalue(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

keyvalue(name="sahil",age=25,number=21314234,gender="M")
```

```python
#combination of both
def comb(*args,**kwargs):
    for i in args:
        print(i)

    for key,value in kwargs.items():
        print(f"{key}:{value}") 

comb(12,435,35,23,4,name="sahil",age=12)
```

```python
def passlen(passw):
    if len(passw)<8:
        return print("Password length should be greater than 8 char")
    if not any(i.isdigit() for i in passw):
        return print("Password should contain a number")
    if not any (i.lower() for i in passw):
        return print("Password should contains lower char")
    if not any (i.upper() for i in passw):
        return print("Password should contains upper char")
    if not any(i in "!@#$%^&*?" for i in passw):
        return print("Password should contains special char")
    return print("Strong Password")
passlen("Aa@")
```

### Imbalanced dataset

```python
import numpy as np 
import pandas as pd
np.random.seed(123)

class0_ratio=0.9
n_sample=1000

n_class0=int(n_sample*class0_ratio)
n_class1= n_sample - n_class0

```

```python
class0=pd.DataFrame({
  "Feature 1": np.random.normal(0,1,n_class0),
  "Feature 2": np.random.normal(0,1,n_class0),
  "Target": [0]*n_class0
})

class1=pd.DataFrame({
  "Feature 1": np.random.normal(0,1,n_class1),
  "Feature 2": np.random.normal(0,1,n_class1),
  "Target": [1]*n_class1
})


df=pd.concat((class0,class1)).reset_index(drop=True)
df.head()
```

```python
df["Target"].value_counts()
```

#### upsampling the data

```python
df_minority=df[df["Target"]==1]

df_majority=df[df["Target"]==0]
```

## Handling imbalanced dataset with SMOTE 

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_features=2,weights=[0.90],random_state=12,n_redundant=0,n_samples=1000,n_clusters_per_class=1)
x,y
```

```python
feature=pd.DataFrame(x,columns=['f1','f2'])
target=pd.DataFrame(y,columns=['Target'])

df=pd.concat([feature,target],axis=1)
df.head()
```

```python
df.shape
```

```python
df.Target.value_counts()
```

```python
plt.scatter(df['f1'],df['f2'],c=df['Target'])
```

```python
from imblearn.over_sampling import SMOTE

oversample=SMOTE()

x,y=oversample.fit_resample(df[['f1','f2']],df['Target'])
```

```python
df_oversampled=pd.concat([x,y],axis=1).reset_index(drop=True)

df_oversampled.shape
```

```python
df_oversampled.Target.value_counts()
```

```python
plt.scatter(df_oversampled['f1'],df_oversampled['f2'],c=df_oversampled['Target'])
plt.show()
```

## Outliers 

```python
lst=[-200,-23,32,23,6,23,64,3,32,45,34,7,7,6786,56,45,34,76,65,44,56,47,75,88,100,56,84,43,77,89]

min,q1,mediann,q3,max=np.quantile(lst,[0,0.25,0.50,0.75,1])

min,q1,mediann,q3,max
```

```python
IQR=q3-q1
IQR
```

```python
lower=q1-1.5*IQR
higher=q3+1.5*IQR

lower,higher
```

```python
sns.boxplot(lst)
plt.title("With outliers")
```

```python
#finding outline in the lst 
ls=[i for i in lst if i>higher or i<lower]
ls
```

```python
#removing outliers
cleaned_lst=[i for i in lst if i not in ls]
cleaned_lst
```

```python
sns.boxplot(cleaned_lst)
plt.title("After removing outliers")
```

### Target Guided Encoding
It is used when we have multiple unique categories in column so we encode on the basis of the target value. 
We take the mean of the target value for that specific category and replace it with the mean or median

```python
df=pd.DataFrame({
    "City":["New York","Delhi","Mumbai","Delhi","Kolkata","New York","Mumbai","Pune","Pune","Dwarka"],
    "Price":[200,121,354,232,532,220,564,333,224,465]
})
df
```

```python
city_mean=df.groupby(["City"])["Price"].mean().to_dict()
```

```python
city_mean
```

```python
df["City_encoded"]=df["City"].map(city_mean)

df
```
