```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

### DataSet

```python
x= 6* np.random.randn(500,1) - 3

y=0.5+x**2+1.5 * x +  np.random.randn(500,1)
```

```python
plt.scatter(x,y)
```

### With Linear Regression 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=123,test_size=0.2)
```

```python
model=LinearRegression().fit(x_train,y_train)
```

```python
y_pre=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pre)
mae=mean_absolute_error(y_test,y_pre)
r2=r2_score(y_test,y_pre)
mse,mae,r2
```

```python
plt.plot(x_train,model.predict(x_train),c="r")
plt.scatter(x_train,y_train)
plt.show()
```

### With Polynomial Degree

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
ploy=PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=ploy.fit_transform(x_train)
x_test_poly=ploy.transform(x_test)
```

```python
model2=LinearRegression()
model2.fit(x_train_poly,y_train)
r22=r2_score(y_test,model2.predict(x_test_poly))
r22
```

```python
plt.scatter(x_train,y_train,c='g')
plt.scatter(x_train,model2.predict(x_train_poly),c='r')
plt.legend(["Predicted Line","True Value"])
plt.show()
```
